```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
x_data = [[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]]
y_data = [[0], [0], [0], [1], [1], [1]]

x_train = torch.FloatTensor(x_data)
y_train = torch.FloatTensor(y_data)

print(x_train)
print(y_train)
```

```python
# 시그모이드 함수 정의 

def sigmoid(x):
    return 1/(1+torch.exp(-x))
```

```python

# parameter 정의 h = sigmoid(w*x+b)
W = torch.zeros((2,1),requires_grad=True)
b = torch.zeros(1,requires_grad=True)

# optimizer
optimizer = optim.SGD([W,b],lr=1)


nb_epochs = 20000

for epoch in range(1,nb_epochs+1):
    
    
    Hx = sigmoid(x_train.matmul(W)+b) # 6 x 1 배열 
    
    cost = -(y_train * torch.log(Hx) + (1 - y_train) * torch.log(1 - Hx)).mean()
    cost2 = -torch.mean(y_train*torch.log(Hx)+(1-y_train)*torch.log(1-Hx))
    
    optimizer.zero_grad()
    cost2.backward()
    optimizer.step()
    
    if epoch%100==0:
        print("{}/{} cost: {} ".format(epoch,nb_epochs,cost2.item()))
    
```
